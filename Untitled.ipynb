```python
# Import necessary libraries
from pynq.overlays.base import BaseOverlay
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import erfc
import plotly.graph_objs as go
import plotly.express as px

# Load the base overlay
base = BaseOverlay('base.bit')

# Initialize RF clocks
base.init_rf_clks()

# Initialize the receiver
receiver = base.radio.receiver

# Channel index for ADC A (adjust if necessary)
channel_index = 0  # Assuming channel 0 corresponds to ADC A

# Configure ADC channel settings
receiver.channel[channel_index].adc_block.MixerSettings['Freq'] = 439.1e6  # Set frequency to 439.1 MHz

# Define the number of samples and the sampling frequency
number_samples = 1200  # Number of samples to collect
sample_frequency = 2458e6  # Sample frequency in Hz

# Function to collect data from the specified channel
def collect_data(channel, num_samples):
    """Collects data from the specified channel."""
    return receiver.channel[channel].transfer(num_samples)

# Collect data from ADC A (channel 0)
cdata = collect_data(channel_index, number_samples)

# Function to plot time-domain data
def plot_time_domain(data, fs, title='Time Domain Signal'):
    """Plots the time domain signal.""" 
    n_samples = len(data)
    time_axis = np.linspace(0, n_samples / fs, n_samples, endpoint=False)  # Ensure matching lengths
    plt.figure(figsize=(10, 6))
    plt.plot(time_axis, np.real(data), label='Real Part', color='blue')
    plt.plot(time_axis, np.imag(data), label='Imaginary Part', color='red', linestyle='--')
    plt.title(title)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()
    plt.show()

# Function to plot frequency-domain data (positive frequencies only)
def plot_positive_frequency_domain(data, freq_axis, title='Frequency Domain Signal'):
    """Plots the frequency domain signal (positive frequencies only)."""
    positive_freq_indices = freq_axis > 0  # Select indices where frequency is positive
    plt.figure(figsize=(10, 6))
    plt.plot(freq_axis[positive_freq_indices], np.abs(data)[positive_freq_indices], label='Magnitude', color='blue')
    plt.title(title)
    plt.xlabel('Frequency (MHz)')
    plt.ylabel('Magnitude')
    plt.grid()
    plt.legend()
    plt.show()

# Plot the time-domain data
plot_time_domain(cdata, sample_frequency, title='Time Domain Plot of ADC A Channel')

# Apply a window function (e.g., Blackman) to the data
window = np.blackman(number_samples)
windowed_data = cdata * window

# Perform FFT and shift the frequency domain data
fdata = np.fft.fftshift(np.fft.fft(windowed_data))

# Calculate frequency axis and convert to MHz
freq_axis = np.fft.fftshift(np.fft.fftfreq(len(windowed_data), d=1/sample_frequency)) / 1e6  # Convert to MHz

# Apply a low-pass filter with a cutoff of 120 MHz (only keeping frequencies below 120 MHz)
cutoff_freq = 120  # Cutoff frequency in MHz
filter_mask = np.abs(freq_axis) <= cutoff_freq  # Create a mask to filter frequencies > 120 MHz
filtered_fdata = fdata * filter_mask  # Apply the filter by multiplying

# Plot the filtered frequency-domain data (positive frequencies only)
plot_positive_frequency_domain(filtered_fdata, freq_axis, title='Filtered Frequency Domain Plot (120 MHz Cutoff)')

# Perform inverse FFT to get the filtered time-domain signal
filtered_time_data = np.fft.ifft(np.fft.ifftshift(filtered_fdata))

# Plot the filtered time-domain signal
plot_time_domain(filtered_time_data, sample_frequency, title='Filtered Time Domain Plot of ADC A Channel')

# Now, let's apply Gaussian filters to extract desired frequencies
# Gaussian filter parameters for 103 MHz 
center_freq_103 = 103  # Center frequency in MHz
bandwidth_103 = 0.7  # Bandwidth of the filter in MHz (adjust this based on the width you want around the center frequency)
gaussian_filter_103 = np.exp(-0.5 * ((freq_axis - center_freq_103) / bandwidth_103) ** 2)  # Gaussian filter for 103 MHz

# Apply the Gaussian filter to the previously filtered signal (after the 120 MHz low-pass)
gaussian_filtered_fdata_103 = filtered_fdata * gaussian_filter_103

# Plot the Gaussian filtered frequency-domain data for 103 MHz (positive frequencies only)
plot_positive_frequency_domain(gaussian_filtered_fdata_103, freq_axis, title='Gaussian Filtered Frequency Domain (103 MHz)')

# Perform inverse FFT to get the Gaussian filtered time-domain signal for 103 MHz
gaussian_filtered_time_data_103 = np.fft.ifft(np.fft.ifftshift(gaussian_filtered_fdata_103))

# Plot the Gaussian filtered time-domain signal for 103 MHz
plot_time_domain(gaussian_filtered_time_data_103, sample_frequency, title='Gaussian Filtered Time Domain Plot (103 MHz)')

# Gaussian filter parameters for 89.7 MHz
center_freq_89_7 = 89.7  # Center frequency in MHz
bandwidth_89_7 = 0.7  # Bandwidth of the filter in MHz
gaussian_filter_89_7 = np.exp(-0.5 * ((freq_axis - center_freq_89_7) / bandwidth_89_7) ** 2)  # Gaussian filter for 89.7 MHz

# Apply the Gaussian filter to the previously filtered signal (after the 120 MHz low-pass)
gaussian_filtered_fdata_89_7 = filtered_fdata * gaussian_filter_89_7

# Plot the Gaussian filtered frequency-domain data for 89.7 MHz (positive frequencies only)
plot_positive_frequency_domain(gaussian_filtered_fdata_89_7, freq_axis, title='Gaussian Filtered Frequency Domain (89.7 MHz)')

# Perform inverse FFT to get the Gaussian filtered time-domain signal for 89.7 MHz
gaussian_filtered_time_data_89_7 = np.fft.ifft(np.fft.ifftshift(gaussian_filtered_fdata_89_7))

# Plot the Gaussian filtered time-domain signal for 89.7 MHz
plot_time_domain(gaussian_filtered_time_data_89_7, sample_frequency, title='Gaussian Filtered Time Domain Plot (89.7 MHz)')

# Gaussian filter parameters for 95.5 MHz
center_freq_95_5 = 95.5  # Center frequency in MHz
bandwidth_95_5 = 0.7  # Bandwidth of the filter in MHz
gaussian_filter_95_5 = np.exp(-0.5 * ((freq_axis - center_freq_95_5) / bandwidth_95_5) ** 2)  # Gaussian filter for 95.5 MHz

# Apply the Gaussian filter to the previously filtered signal (after the 120 MHz low-pass)
gaussian_filtered_fdata_95_5 = filtered_fdata * gaussian_filter_95_5

# Plot the Gaussian filtered frequency-domain data for 95.5 MHz (positive frequencies only)
plot_positive_frequency_domain(gaussian_filtered_fdata_95_5, freq_axis, title='Gaussian Filtered Frequency Domain (95.5 MHz)')

# Perform inverse FFT to get the Gaussian filtered time-domain signal for 95.5 MHz
gaussian_filtered_time_data_95_5 = np.fft.ifft(np.fft.ifftshift(gaussian_filtered_fdata_95_5))

# Plot the Gaussian filtered time-domain signal for 95.5 MHz
plot_time_domain(gaussian_filtered_time_data_95_5, sample_frequency, title='Gaussian Filtered Time Domain Plot (95.5 MHz)')
#xxxx

# Function to plot all three Gaussian-filtered frequency-domain data in one plot using Plotly
def plot_all_gaussian_filtered_frequencies(freq_axis, data_103, data_89_7, data_95_5, title='Gaussian Filtered Frequency Domain (All Frequencies)'):
    """Plots all three Gaussian-filtered frequency-domain data in one plot."""
    positive_freq_indices = (freq_axis > 0) & (freq_axis <= 200)  # Select indices where frequency is positive and <= 200 MHz
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=freq_axis[positive_freq_indices], y=np.abs(data_103)[positive_freq_indices], mode='lines', name='103 MHz', line=dict(color='blue')))
    fig.add_trace(go.Scatter(x=freq_axis[positive_freq_indices], y=np.abs(data_89_7)[positive_freq_indices], mode='lines', name='89.7 MHz', line=dict(color='green')))
    fig.add_trace(go.Scatter(x=freq_axis[positive_freq_indices], y=np.abs(data_95_5)[positive_freq_indices], mode='lines', name='95.5 MHz', line=dict(color='red')))
    fig.update_layout(title=title, xaxis_title='Frequency (MHz)', yaxis_title='Magnitude', legend_title='Frequencies', template='plotly_white')
    fig.show()

# Plot all three Gaussian-filtered frequency-domain data in one plot
plot_all_gaussian_filtered_frequencies(freq_axis, gaussian_filtered_fdata_103, gaussian_filtered_fdata_89_7, gaussian_filtered_fdata_95_5)

#xxxxxxxx
# Function to calculate SNR in dB
def calculate_snr(signal, noise):
    """Calculates the SNR (Signal-to-Noise Ratio) in dB.""" 
    signal_power = np.mean(np.abs(signal) ** 2)
    noise_power = np.mean(np.abs(noise) ** 2)
    snr = 10 * np.log10(signal_power / noise_power)
    return snr

# Amplify the Gaussian filtered signals
amplification_factor = 10e4  # Adjust this factor as needed to achieve desired SNR

# Amplify the Gaussian filtered signals
amplified_gaussian_filtered_time_data_103 = gaussian_filtered_time_data_103 * amplification_factor
amplified_gaussian_filtered_time_data_89_7 = gaussian_filtered_time_data_89_7 * amplification_factor
amplified_gaussian_filtered_time_data_95_5 = gaussian_filtered_time_data_95_5 * amplification_factor

# Introduce noise to the signals to calculate SNR
# Generate random noise with same shape as the signals
noise_level = 0.05  # Adjust this value to control the amount of noise
noise_103 = noise_level * np.random.randn(len(amplified_gaussian_filtered_time_data_103))
noise_89_7 = noise_level * np.random.randn(len(amplified_gaussian_filtered_time_data_89_7))
noise_95_5 = noise_level * np.random.randn(len(amplified_gaussian_filtered_time_data_95_5))

# Add noise to the amplified signals
noisy_signal_103 = amplified_gaussian_filtered_time_data_103 + noise_103
noisy_signal_89_7 = amplified_gaussian_filtered_time_data_89_7 + noise_89_7
noisy_signal_95_5 = amplified_gaussian_filtered_time_data_95_5 + noise_95_5

# Calculate SNR for each signal
snr_103 = calculate_snr(amplified_gaussian_filtered_time_data_103, noise_103)
snr_89_7 = calculate_snr(amplified_gaussian_filtered_time_data_89_7, noise_89_7)
snr_95_5 = calculate_snr(amplified_gaussian_filtered_time_data_95_5, noise_95_5)

# Print the SNR values
print(f"SNR for 103 MHz: {snr_103:.2f} dB")
print(f"SNR for 89.7 MHz: {snr_89_7:.2f} dB")
print(f"SNR for 95.5 MHz: {snr_95_5:.2f} dB")


# Convert SNR from dB to linear
snr_103_linear = 10 ** (snr_103 / 10)
snr_89_7_linear = 10 ** (snr_89_7 / 10)
snr_95_5_linear = 10 ** (snr_95_5 / 10)

# Calculate BER using Q-function for BPSK
def calculate_ber(snr_linear):
    return 0.5 * erfc(math.sqrt(2 * snr_linear))

ber_103 = calculate_ber(snr_103_linear)
ber_89_7 = calculate_ber(snr_89_7_linear)
ber_95_5 = calculate_ber(snr_95_5_linear)

# Print the BER values
print(f"BER for 103 MHz: {ber_103:.10f}")
print(f"BER for 89.7 MHz: {ber_89_7:.10f}")
print(f"BER for 95.5 MHz: {ber_95_5:.10f}")


#SNR and Ber
snr_values_db = [snr_103, snr_89_7, snr_95_5]
ber_values = [ber_103, ber_89_7, ber_95_5]

frequencies = ['103 MHz', '89.7 MHz', '95.5 MHz']

# Custom frequency labels for the stations
frequency_labels = ['Antenne Bayern (103 MHz)', 'Deutschlandfunk (95.5 MHz)', 'Gong FM (89.7 MHz)']

# Plotting SNR values in dB
plt.figure(figsize=(10, 6))
plt.bar(frequency_labels, snr_values_db, color='blue')
plt.xlabel('Frequency')
plt.ylabel('SNR (dB)')
plt.title('SNR Values in dB')
plt.grid(True)
plt.show()

# Plotting BER values
plt.figure(figsize=(10, 6))
plt.bar(frequency_labels, ber_values, color='red')
plt.xlabel('Frequency')
plt.ylabel('BER')
plt.title('BER Values')
plt.yscale('log')  # Logarithmic scale for BER values
plt.grid(True)
plt.show()
```
