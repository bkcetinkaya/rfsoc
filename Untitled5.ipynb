```python
# Import necessary libraries
from pynq.overlays.base import BaseOverlay
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import erfc
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd  # For saving to Excel

# Load the base overlay
base = BaseOverlay('base.bit')

# Initialize RF clocks
base.init_rf_clks()

# Initialize the receiver
receiver = base.radio.receiver

# Channel index for ADC A (adjust if necessary)
channel_index = 0  # Assuming channel 0 corresponds to ADC A

# Configure ADC channel settings
receiver.channel[channel_index].adc_block.MixerSettings['Freq'] = 439.1e6  # Set frequency to 439.1 MHz

# Define the number of samples and the sampling frequency
number_samples = 4800  # Number of samples to collect
sample_frequency = 245e6  # Sample frequency in Hz

# Function to collect data from the specified channel
def collect_data(channel, num_samples):
    """Collects data from the specified channel."""
    return receiver.channel[channel].transfer(num_samples)

# Collect data from ADC A (channel 0)
cdata = collect_data(channel_index, number_samples)

# Save the data to an Excel file
def save_data_to_excel(data, filename="collected_data2.xlsx"):
    """Saves the collected data to an Excel file."""
    df = pd.DataFrame({
        'Real Part': np.real(data),
        'Imaginary Part': np.imag(data)
    })
    df.to_excel(filename, index=False)
    print(f"Data saved to {filename}")

# Save the collected data
save_data_to_excel(cdata)

# Function to plot time-domain data
def plot_time_domain(data, fs, title='Time Domain Signal'):
    """Plots the time domain signal."""
    n_samples = len(data)
    time_axis = np.linspace(0, n_samples / fs, n_samples, endpoint=False)  # Ensure matching lengths
    plt.figure(figsize=(10, 6))
    plt.plot(time_axis, np.real(data), label='Real Part', color='blue')
    plt.plot(time_axis, np.imag(data), label='Imaginary Part', color='red', linestyle='--')
    plt.title(title)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()
    plt.show()

# Plot the time-domain data
plot_time_domain(cdata, sample_frequency, title='Time Domain Plot of ADC A Channel')
```
